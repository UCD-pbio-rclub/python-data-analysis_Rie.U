## question 1

```python
weather = pd.read_csv('https://raw.githubusercontent.com/hadley/nycflights13/master/data-raw/weather.csv')
```

```python
weather.head()
```

```python
weatherTimeCourse = weather.groupby(['origin','month'])[['temp','humid']].mean()
```

```python
weatherTimeCourse
```

```python
EWR = weatherTimeCourse.loc['EWR']
```

```python
JFK = weatherTimeCourse.loc['JFK']
```

```python
LGA = weatherTimeCourse.loc['LGA']
```

```python
fig,axes = plt.subplots(3,1, sharex = True)
```

```python
EWR.plot(ax=axes[0], title = 'EWR', use_index=True)
```

```python
JFK.plot(ax=axes[1], title = 'JFK', use_index=True)
```

```python
LGA.plot(ax=axes[2], title = 'LGA', use_index=True)
```

```python
plt.close()
```

```python
humid = weather.groupby(['origin','month','day'])[['humid']].mean()
```
