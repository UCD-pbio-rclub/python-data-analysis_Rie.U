```python
population = pd.read_csv("https://raw.githubusercontent.com/jakevdp/data-USstates/master/state-population.csv")
areas = pd.read_csv("https://raw.githubusercontent.com/jakevdp/data-USstates/master/state-areas.csv")
stateabrevs = pd.read_csv("https://raw.githubusercontent.com/jakevdp/data-USstates/master/state-abbrevs.csv")
```

```python
population.head()
```

```python
areas.head()
```

```python
stateabrevs.head()
```

```python
abbAreas = pd.merge(areas, stateabrevs, on='state', how='outer')
```

```python
abbAreas.head()
```

```python
allComb = pd.merge(population, abbAreas, left_on='state/region', right_on='abbreviation', how='outer')
```

```python
allComb.head()
```

```python
allComb['density'] = allComb['population']/allComb['area (sq. mi)']
```

```python
allComb.head()
```

```python
allComb2 = allComb[['state/region','ages', 'density']]
```

```python
allComb2.head()
```

```python
allComb3 = allComb2[allComb2['ages'].str.contains('total')==True]
```

```python
allComb3.head()
```
